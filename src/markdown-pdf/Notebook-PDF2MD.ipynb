```python

import os
import fitz  # PyMuPDF library for working with PDF files
# Define the file path in the fabric lakehouse
lakehouse_bronze_name = "lh_adventureworks_bronze"
folder_name = "policydocuments"
lakehouse_silver_name = "lh_adventureworks_silver"

def pdf_to_markdown(pdf_file_path, markdown_file_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_file_path)
    
    markdown_text = ""
    
    for page_num in range(pdf_document.page_count):
        page = pdf_document[page_num]
        markdown_text += page.get_text("text")
    
    # Write the extracted text to a Markdown file
    with open(markdown_file_path, "w") as markdown_file:
        markdown_file.write(markdown_text)




# Define the file path in the fabric lakehouse
file_path_bronze = f"lakehouse/{lakehouse_bronze_name}/Files/{folder_name}"
file_path_silver = f"lakehouse/{lakehouse_silver_name}/Files{folder_name}"

# Loop over all files in the fabric lakehouse
for file_name in os.listdir(file_path_bronze):
    if file_name.endswith(".pdf"):
        print(file_name)
        pdf_to_markdown(os.path.join(file_path_bronze, file_name), os.path.join(file_path_silver, replace(file_name, ".pdf", ".md")))
```
